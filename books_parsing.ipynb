```python
import csv
import urllib.request
from bs4 import BeautifulSoup
import re
import datetime

BASE_URL = 'https://book24.ru/catalog/fiction-1592/page-'
site = 'https://book24.ru'
content = {}
genres = []

def get_html(url):
    response = urllib.request.urlopen(url)
    return response.read()

def parse_book(html):
    soup = BeautifulSoup(html)
    div = soup.find('div',{'class' : 'catalog-products__list'}).findAll('a', {'class' : 'book__title-link'})
 
    with open('out_books_genres.txt','w') as out:
        for i in div:
            genre = parse_genre(get_html(site + i.get('href')))
            autor = parse_autor(get_html(site + i.get('href')))
            content[i.string.strip()] = genre
            print('{}:{}:{}'.format(autor, i.string.strip(), genre))
            out.write('{}:{}:{}\n'.format(autor, i.string.strip(), genre))
    
def parse_genre(html):
    soup = BeautifulSoup(html)
    genre = soup.find(text = re.compile('Раздел:'))
    try:
        test = genre.next
        test = test.next
    except AttributeError:
        test = '-'
        return test
    #finally:
    else:
        return test.text.strip()

def parse_autor(html):
    soup = BeautifulSoup(html)
    autor = soup.find(text = re.compile('Автор:'))
    try:
        test = autor.next
        test = test.next
    except AttributeError:
        test = '-'
        return test
    #finally:
    else:
        return test.text.strip()
    
def main():
    i = 0
    while(i < 431):
        i += 1
        parse_book(get_html(BASE_URL + str(i)))
        with open('out_page_number.txt','w') as out:
            out.write('{}\n'.format(i))
        print('{} - {}\n'.format(i, datetime.datetime.now()))
            
if __name__ == '__main__':
    main()
```
