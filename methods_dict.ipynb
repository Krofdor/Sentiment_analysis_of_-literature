```python
from sklearn.metrics import classification_report
from matplotlib import pylab
from collections import Counter

class Labeled_Words:
    def __init__(self, words, targets):##, target_names):
        self.words = words
        self.targets = targets
##        self.target_names = target_names
        
    def __eq__(self, other):
        if isinstance(other, Number):
            return self.number == other.number
        return False

##########################################

list_of_Labeled_words = Labeled_Words([],[])
list_of_Labeled_words.words = all_classes_df['Russian Translation (Google Translate)']
list_of_Labeled_words.targets = all_classes_df['Class']

##########################################

pylab.figure(figsize=(10,6))
pylab.bar(Counter(list_of_Labeled_words.targets).keys(), Counter(list_of_Labeled_words.targets).values())

print("Точность на обучающей выборке:\n")

docs_test = list_of_Labeled_words.words

nb_clf = Pipeline([('vect', CountVectorizer()),
                     ('tfidf', TfidfTransformer()),
                     ('mnb', MultinomialNB()),
                    ])

nb_clf = nb_clf.fit(list_of_Labeled_words.words, list_of_Labeled_words.targets)

predicted = nb_clf.predict(docs_test)
nb = np.mean(predicted == list_of_Labeled_words.targets)
print("CV + TF-IDF + NB: ", nb) 
print(classification_report(predicted, list_of_Labeled_words.targets))

svm_clf = Pipeline([('vect', CountVectorizer()),
                     ('tfidf', TfidfTransformer()),
                     ('svm', SGDClassifier(loss='hinge', penalty='l2', alpha=1e-3, n_iter=5, random_state=42))
                    ])

svm_clf = svm_clf.fit(list_of_Labeled_words.words, list_of_Labeled_words.targets)

predicted = svm_clf.predict(docs_test)
svm = np.mean(predicted == list_of_Labeled_words.targets)
print("CV + TF-IDF + SVM: ", svm) 
print(classification_report(predicted, list_of_Labeled_words.targets))

knn_clf = Pipeline([('vect', CountVectorizer()),
                     ('tfidf', TfidfTransformer()),
                     ('knn', KNeighborsClassifier(n_neighbors=3, p=2, algorithm='kd_tree'))
                    ])

knn_clf = knn_clf.fit(list_of_Labeled_words.words, list_of_Labeled_words.targets)

predicted = knn_clf.predict(docs_test)
knn = np.mean(predicted == list_of_Labeled_words.targets)
print("CV + TF-IDF + k-NN: ", knn) 
print(classification_report(predicted, list_of_Labeled_words.targets))

##########################################

docs_test = words_with_emotions['Russian']
#docs_test = all_classes_df['Russian Translation (Google Translate)']

predicted = nb_clf.predict(docs_test)
#nb = np.mean(predicted == list_of_Labeled_words.targets)
print("CV + TF-IDF + NB: ", predicted) 

predicted = svm_clf.predict(docs_test)
#svm = np.mean(predicted == list_of_Labeled_words.targets)
print("CV + TF-IDF + SVM: ", predicted) 

predicted = knn_clf.predict(docs_test)
#knn = np.mean(predicted == list_of_Labeled_words.targets)
print("CV + TF-IDF + k-NN: ", predicted)

##########################################

#Визуализация результата

%matplotlib inline

emotions_score = []
binary_score = []

a= binary_score.append(words_with_emotions.Positive.sum())
b= binary_score.append(words_with_emotions.Negative.sum())

c= emotions_score.append(words_with_emotions.Anger.sum())
d= emotions_score.append(words_with_emotions.Anticipation.sum())
e= emotions_score.append(words_with_emotions.Disgust.sum())
f= emotions_score.append(words_with_emotions.Fear.sum())
g= emotions_score.append(words_with_emotions.Joy.sum())
h= emotions_score.append(words_with_emotions.Sadness.sum())
m= emotions_score.append(words_with_emotions.Surprise.sum())
n= emotions_score.append(words_with_emotions.Trust.sum())

binary_score_df = pd.DataFrame(pd.np.empty((0, 2)))
binary_score_df.columns = words_with_emotions.columns[1:3]
binary_score_df.loc[0] = [x for x in binary_score]
print(binary_score_df)

emotions_score_df = pd.DataFrame(pd.np.empty((0, 8)))
emotions_score_df.columns = words_with_emotions.columns[3:11]
emotions_score_df.loc[0] = [x for x in emotions_score]
print(emotions_score_df)
```
